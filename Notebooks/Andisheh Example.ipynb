```python
import os, datetime
import pandas as pd
from fasttrips import Run

BASE_DIR = "C:/apps/fast-trips"
INPUT_NETWORKS = os.path.join(BASE_DIR, 'fasttrips', 'Examples', 'networks', 'flexpress')
INPUT_DEMAND = os.path.join(BASE_DIR, 'fasttrips', 'Examples', 'flexpress')
INPUT_WEIGHTS = os.path.join(INPUT_DEMAND, 'pathweight_ft.txt')
RUN_CONFIG = os.path.join(INPUT_DEMAND, 'config_ft.txt')
OUTPUT_DIR = INPUT_DEMAND
OUTPUT_FOLDER = os.path.join(OUTPUT_DIR, 'output')

def run_fasttrips():
    Run.run_fasttrips(
        input_network_dir = INPUT_NETWORKS,
        input_demand_dir = INPUT_DEMAND,
        run_config = RUN_CONFIG,
        input_weights = INPUT_WEIGHTS,
        output_dir = OUTPUT_DIR,
        output_folder = OUTPUT_FOLDER,
        pathfinding_type = 'stochastic',
        iters = 1,
        overlap_variable = "None",
        dispersion = 1.0
    )

if __name__ == '__main__':
    run_fasttrips()
```
